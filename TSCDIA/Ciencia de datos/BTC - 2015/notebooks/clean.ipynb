## importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Route of csv info
path_csv = '../datos/brutos/BTC.csv'
```

```python
#read of csv data
datos = pd.read_csv(path_csv, sep=',', encoding='latin1')
```

```python
datos.sample(2, random_state=3)
```

```python
# Convert row 'Date' to type date time
datos.Date = pd.to_datetime(datos.Date)
```

## opening prices in 2015

```python
# Filter rows of dataset by year of date
datos_2015 = datos.loc[datos['Date'].dt.year == 2015]
```

```python
#Stats by data filtred
datos_2015.describe()
```

```python
# Chart of Bitcoin opening Prices in 2015
plt.plot(datos_2015.Date, datos_2015.Open)
plt.axhline(datos_2015.Open.mean(), color='red', linestyle='--', label='Mean')
plt.title('Variability of opening price at BTC in 2015')
plt.xlabel('date yer/mont/day')
plt.ylabel('price values of BTC')
plt.legend()  # Agregar leyenda si es necesario
plt.show()
```

```python
plt.bar(datos_2015.Date.dt.month, datos_2015.Open)
plt.title('Values opening priece for months in 2015')
plt.xlabel('numbers of months')
plt.ylabel('Prices')
```

## Closing prices in 2015

```python
#chart of bitcoin Closing prices in 2015
plt.plot(datos_2015.Date, datos_2015.Close)
plt.axhline(datos_2015.Close.mean(), color='red', linestyle='--', label='Media')
plt.title('Closing price values at BTC in 2015')
plt.xlabel('Closing values prices')
plt.ylabel('date yer/mont/day')
```

```python
plt.bar(datos_2015.Date.dt.month, datos_2015.Close)
plt.title('Values closing prieces for months in 2015')
plt.xlabel('Months')
plt.ylabel('Prices')
```
