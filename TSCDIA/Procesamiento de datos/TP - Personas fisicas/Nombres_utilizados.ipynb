```python
import pandas as pd
import matplotlib.pyplot as plt
```

Link de dataset: https://www.kaggle.com/datasets/akielbowicz/nombres-de-personas-fsicas-de-argentina?select=historico-nombres.csv

```python
# Cargar el archivo en un DataFrame
df = pd.read_csv('./Datos/historico-nombres.csv')

# Mostrar las primeras 5 filas del DataFrame
print(df.head())
```

```python
# Consulta: Mostrar la cantidad total de registros en el DataFrame
total_registros = len(df)
print("Cantidad total de registros:", total_registros)

# Consulta: Calcular la cantidad total de nombres únicos
nombres_unicos = df['nombre'].nunique()
print("Cantidad total de nombres únicos:", nombres_unicos)

# Consulta: Calcular la cantidad total de registros por año
registros_por_anio = df.groupby('anio')['cantidad'].sum()

# Gráfico: Visualizar la cantidad de registros por año
plt.figure(figsize=(10, 6))
registros_por_anio.plot(kind='line')
plt.title('Cantidad de Registros por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Solicitar al usuario que ingrese el nombre a consultar
nombre_consulta = input("Ingrese el nombre a consultar: ")
# Filtrar los registros con el nombre ingresado
registros_nombre = df[df['nombre'] == nombre_consulta]
```

```python
# Obtener la cantidad de registros del nombre ingresado
cantidad_registros = len(registros_nombre)
print("Cantidad de años que hay registros para el nombre", nombre_consulta + ":", cantidad_registros)

# Obtener los años únicos en los que se registraron personas con el nombre ingresado
anios_registro = registros_nombre['anio'].unique()

# Mostrar los años en los que se registraron personas con el nombre ingresado
print("Años en los que se registraron personas con el nombre", nombre_consulta + ":")
for anio in anios_registro:
    print(anio)

#Calcular la suma de los registros de los años en que aparece el nombre ingresado
print('Cantidad de años en los que aparece el nombre: ', len(anios_registro))
```
