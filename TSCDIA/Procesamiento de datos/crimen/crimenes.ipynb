## ETL

```python
#Renombrando nombre de columnas
datos = datos.rename(columns={'Sexo/Averiguaciï¿½n previa':'sexo',
                        'Bien jurï¿½dico afectado':'Bien afectado'})
```

## Despivoteo de la tabla por medio de melt

Las columnas de los meses pasan como datos a una sola columna de "Mes"

```python
valores = [col for col in datos.columns if col !='Clave_Ent' and col !='Entidad' and col != 'Bien afectado' and col != 'Tipo de delito' and col != 'Subtipo de delito' and col != 'Modalidad' and col != 'sexo' and col != 'Rango de edad']
```

```python
datos = pd.melt(datos, id_vars = ['Clave_Ent', 'Entidad', 'Bien afectado', 'Tipo de delito', 'Subtipo de delito', 'Modalidad', 'sexo', 'Rango de edad'], value_vars = valores,  var_name = 'mes' )
```

### Eliminacion de datos nulos, duplicados y mal tipeados

```python
datos = datos.dropna()
```

```python
datos = datos.drop_duplicates()
```

```python
filtro = (datos['sexo'] != 'Mujer')
```

```python
filtro_dos = (datos['sexo'] != 'Hombre')
```

```python
datos['sexo'] = datos['sexo'] != filtro
```

```python
datos['sexo'] = datos['sexo'] != filtro
```
