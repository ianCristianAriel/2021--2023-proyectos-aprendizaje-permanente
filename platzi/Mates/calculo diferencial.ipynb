```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#f(x) = x **2
def f(x):
    return x ** 2

x = np.linspace(-20, 20, num=50)
y = f(x)
```

```python
#Derivada -> F(x) -> x ** 2
def f_d(f):
    h = 0.00002
    return (f(x+h) - f(x)) / h
```

```python
plt.plot(x, y, 'b', label='F(x) -> x ** 2')
plt.plot(x, f_d(f), 'r', label='Derivada -> F(x) -> x ** 2')
plt.legend()
plt.title('Funcion y su derivada')
plt.grid()
```
