#Ejercicio 1: Gráficos de Barras

Dibujar un Gráficos de Barras con los siguiente

|-------------|------------|

|Arándano	|100|
|Kiwi     |30|

````python
import pandas as pd
````

````python
import matplotlib.pyplot as plt

fruta ={'manzana':40, 
        'arandano':100, 
        'kiwi':30, 
        'naranja':55}

# Desempaquetar el diccionario en listas separadas para x e y
fruta_names = list(fruta.keys())
fruta_values = list(fruta.values())

# Crear la visualización de barras
plt.bar(fruta_names, fruta_values)
plt.title('Valores de la fruta')
plt.xlabel('Frutas')
plt.ylabel('Valores')
plt.show()
````

#Ejercicio 4: Boxplot



##Boxplot a partir de datos
Realizar un box plot a partir del siguiente dataset [35, 25, 25, 15]

````python
def g_caja(datos, titulo):
    plt.boxplot(datos)
    plt.legend()
    plt.title(titulo)
    plt.show()
````

````python
g_caja(datos, 'Grafico de boxplot, array')
````

## Comparacion de 2 boxplot

Crear dos conjuntos con 200 datos usando `random.normal()` de la libreria numpy. El primer conjunto para una media 100 y desviacion estandar 2 y el segundo conjunto media 100 y desviacion estandar 0.2. Graficar los boxblot y comentar las diferencias.

````python
def boxplot(datos_uno, datos_dos):
    plt.boxplot([datos_uno, datos_dos], labels=['MEDIA: 100 | STD: 2', 'MEDIA: 100 | STD: 0.2'])
    plt.title('Par de boxplot con distribucion normal')
    plt.show()
````

````python
np.random.seed(42)
datos_uno = np.random.normal(100, 2, size=200)
datos_dos = np.random.normal(100, 0.2, size=200)
````

````python
boxplot(datos_uno, datos_dos)
````
