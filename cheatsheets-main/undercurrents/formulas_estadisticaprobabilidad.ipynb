## Estadísticas Descriptivas
- **Aclaraciones generales:**
  - Las probabilidades pueden expresarse tanto en forma de fracción como en porcentaje.
  - En situaciones de eventos sin remplazo, el número de sucesos posibles en iteraciones posteriores se reduce en función de los sucesos previamente tomados.
  - La representación de cualquier probabilidad simple puede variar en complejidad dependiendo del contexto en el que se aplique. por ejemplo la probabilidad conjunta puede ser utilizada de la siguiente manera: ```P(A ∩ B) = P(A) * P(B|A) ```

- **Media (Promedio):** ``` x̄ = (1/n) * Σ(xi) ```

- **Mediana:**
  - Si \(n\) es impar: Mediana = ``` x((n+1)/2)```
  - Si \(n\) es par: Mediana = ``` (x(n/2) + x(n/2 + 1))/2 ```

- **Varianza:** σ² = ``` (1/n) * Σ(xi - x̄)² ```

- **Desviación Estándar:** ``` σ = √(σ²) ```

## Probabilidad

- **Probabilidad de un Evento:** ``` P(A) = (Número de casos favorables) / (Número total de casos posibles) ```

- **Probabilidad Conjunta:** ``` P(A ∩ B) = P(A) * P(B) ```

- **Probabilidad de la unión:** ``` P(A ∪ B) = P(A) + P(B) - P(A ∩ B) ```

- **Probabilidad Condicional:** ``` P(A|B) = P(A ∩ B) / P(B) ```

- **Regla de Bayes:** ``` P(B|A) = [P(A|B) * P(B)] / P(A) ```

## Distribuciones de Probabilidad

- **Distribución Normal:** ``` f(x) = (1 / (σ√(2π))) * e^(-(x - μ)² / (2σ²)) ```

- **Distribución Binomial:** ``` P(X = k) = (n choose k) * p^k * (1-p)^(n-k) ```

- **Distribución Poisson:** ``` P(X = k) = (e^(-λ) * λ^k) / k! ```

## Estimación de Parámetros

- **Estimación de Media Poblacional:** ``` x̂ = x̄ ```

- **Estimación de Varianza Poblacional:** ``` σ̂² = (1/n) * Σ(xi - x̄)² ```
