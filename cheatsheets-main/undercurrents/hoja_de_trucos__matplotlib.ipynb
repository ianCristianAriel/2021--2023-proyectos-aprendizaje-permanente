3. Gráficos:

  *  Grafico de linea:
    *   `plt.plot(x, y, 'color', linewidth=2, linestyle='--', marker='o', markersize=5, label='etiqueta')`

  *  Grafico de dispersion:
    *  `plt.scatter(x, y, color='red', marker='o', label='etiqueta')`

  *  Grafico de barras:
    *  `plt.bar(x, y, color='blue', label='etiqueta')`

  *  Grafico de barras horizontal:
    *  `plt.barh(x, y, color='green', label='etiqueta')`

  *  Grafico de histograma:
    *  `plt.hist(data, bins=10, color='purple', alpha=0.5, label='etiqueta')`

  *  Grafico de torta:
    *   `plt.pie(x, labels=labels, colors=colors, autopct='%1.1f%%')`

  *  Gráficos 3D:
    *   `from mpl_toolkits.mplot3d import Axes3D`
    *   `fig = plt.figure()`
    *   `ax = fig.add_subplot(111, projection='3d')`
    *   `ax.scatter(x, y, z, c='r', marker='o')`

  *  Mapas de calor (heatmaps):
    *   `plt.imshow(data, cmap='hot', interpolation='nearest')`
    *  `plt.colorbar()`

  *  Diagramas de caja (boxplots):
    *   `plt.boxplot(data)`

  *  Gráficos de contorno (contour plots):
    *   `plt.contour(X, Y, Z)`

4. Subplots (gráficos múltiples):

    *  ```
        fig, ax = plt.subplots() # Crear una figura y ejes vacíos
        ax.plot(x, y) # Graficar en los ejes
        ax.set_xlabel('Eje X') # Etiqueta del eje x
        ax.set_ylabel('Eje Y') # Etiqueta del eje y
        ax.set_title('Título del gráfico') # Título del gráfico
        plt.show() # Mostrar el gráfico
       ```
