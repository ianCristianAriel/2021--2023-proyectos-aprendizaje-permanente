### Proiedad asociativa del producto interno

```python
A = np.array([1,2,4])
B = np.array([2,3,5])
C = np.array([3,4,6])
```

```python
# A * (B * C) = (A * B) * C
A.dot(B.dot(C))
```

```python
A.dot(C).dot(B)
```

### Propiedad distributiva del producto interno

```python
# A * ( B + C ) = A * B + A * C
A.dot(B + C)
```

```python
A.dot(B) + A.dot(C)
```

### Proppiedad comutativa del producto interno

```python
# B * C = C * B
B.dot(C)
```

```python
C.dot(B)
```
