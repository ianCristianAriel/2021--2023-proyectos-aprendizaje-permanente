también?

*   c) ¿Cuál es la probabilidad de que la segunda sea roja dado que la
primera resultó blanca?

```python
R = 5
B = 10
S = R + B
```

```python
#a) Teorema de probabilidad total

a = (R / S) * ((R-1)/(S-1))
a
```

```python
#b Teorema de la probabilidad total
# P(R1, R2) + P(B1, R2)

b = (R/S) * ((R-1)/(S-1)) + (B/S) * ((R)/(S-1))
```
