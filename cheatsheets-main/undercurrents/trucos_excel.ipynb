## Funciones Básicas
- `SUMA`: Suma una serie de números.
- `PROMEDIO`: Calcula el promedio de una serie de números.
- `MAX`: Encuentra el valor máximo en una serie de números.
- `MIN`: Encuentra el valor mínimo en una serie de números.
- `CONTAR`: Cuenta la cantidad de celdas que contienen números.
- `CONTARA`: Cuenta la cantidad de celdas que contienen texto.
- `CONTAR.SI`: Cuenta celdas que cumplen una condición específica.
- `SUMAR.SI`: Suma celdas que cumplen una condición específica.
- `PROMEDIO.SI`: Calcula el promedio de celdas que cumplen una condición.
- `SI`: Devuelve un valor según una condición.
- `Y`: Devuelve VERDADERO si todos los argumentos son VERDADERO.
- `O`: Devuelve VERDADERO si al menos un argumento es VERDADERO.
- `NO`: Invierte el valor lógico de una celda.

## Funciones de Búsqueda y Referencia
- `BUSCARV`: Busca un valor en una tabla y devuelve un valor relacionado.
- `BUSCARH`: Busca un valor en una fila y devuelve un valor relacionado.
- `INDICE` y `COINCIDIR`: Buscan y recuperan valores en una matriz.
- `DESREF`: Devuelve una referencia basada en una posición relativa.

## Funciones de Texto
- `CONCATENAR`: Combina el contenido de celdas en una.
- `IZQUIERDA`: Extrae caracteres desde el principio de una celda.
- `DERECHA`: Extrae caracteres desde el final de una celda.
- `MAYUSC`: Convierte texto en mayúsculas.
- `MINUSC`: Convierte texto en minúsculas.
- `NOMPROPIO`: Convierte a mayúsculas la primera letra de palabras.

## Formato Condicional
- Resalta celdas según condiciones para mejorar la visualización.

## Tablas Dinámicas
- Resumen y análisis interactivo de grandes cantidades de datos.

## Gráficos
- Visualización clara y comprensible de datos.

## Validación de Datos
- Restringe el ingreso de datos en celdas, define listas desplegables, etc.

## Filtros
- Muestra solo datos relevantes según criterios.

## Funciones de Fecha y Hora
- `HOY` y `AHORA`: Fechas y horas actuales.
- `DIA`, `MES` y `AÑO`: Extraen partes de una fecha.
- `DIAS360`: Diferencia entre fechas basada en un año de 360 días.

## Funciones Matemáticas Avanzadas
- `POTENCIA`: Eleva un número a una potencia.
- `REDONDEAR`: Redondea un número a decimales.
- `ENTERO`: Redondea un número hacia abajo.
- `ABS`: Valor absoluto de un número.
- `RAIZ`: Raíz cuadrada de un número.

## Funciones Financieras
- `TASA`: Calcula la tasa de interés.
- `PAGO`, `VF`, `VA`: Cálculos financieros.

## Funciones de Base de Datos
- `DGET`, `DSUM`, `DPROMEDIO`: Análisis de bases de datos.
